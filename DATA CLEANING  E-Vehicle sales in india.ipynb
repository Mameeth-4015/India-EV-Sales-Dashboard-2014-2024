```python
df.drop_duplicates(inplace=True)    
```

```python
df["Year"].unique()
```

```python
df["Year"] = df['Year'].replace(".0", "", regex=True).astype(int)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["Month_Name"] = df["Month_Name"].str.title()
```

```python
Category_column = ['Month_Name', 'State', 'Vehicle_Class', 'Vehicle_Category', 'Vehicle_Type']
df[Category_column] = df[Category_column].astype('category')
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['EV_Sales_Quantity']=df['EV_Sales_Quantity'].astype(int)
```

```python
df.info()
```

```python
categorical_columns = ['Month_Name', 'State', 'Vehicle_Class','Vehicle_Category', 'Vehicle_Type']
df[categorical_columns] =df[categorical_columns].astype('category')
```

```python
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df['Sales_Rolling_3'] = df.groupby('State')['EV_Sales_Quantity'].rolling(window=3).mean().reset_index(0, drop=True)
```

```python
df.info()
```

```python
plt.figure(figsize=(6,4))
plt.title('Yearly EV Sales Quantity')
sns.lineplot(x='Year', y='EV_Sales_Quantity', data=df,marker='o', color='black')
plt.xlabel('Year')
plt.ylabel('EV Sales');
```

```python
plt.figure(figsize=(6,4))
plt.title('Monthly EV Sales Quantity')
sns.lineplot(x='Month', y='EV_Sales_Quantity', data=df,marker='o', color='g')
plt.xlabel('Month')
plt.ylabel('EV Sales');
```

```python
plt.figure(figsize=(6,4))
plt.title("Vehicle_Type-wise EV Sales Quantity")
sns.scatterplot(x='Vehicle_Type', y='EV_Sales_Quantity', data=df, color='blue')
plt.xlabel('Vehicle Type')
plt.ylabel('EV Sales')
```

```python
import plotly.express as px

fig = px.line(df, x='Year', y='EV_Sales_Quantity', 
              title='Yearly Analysis Sales in India',
              markers=True,
              labels={'Year': 'Year', 'EV_Sales_Quantity': 'EV Sales'},
              line_shape='linear')

fig.show()
```

```python
plt.figure(figsize=(6,7))
plt.title('State-Wise Analysis of EV Sales')
sns.barplot(y='State', x='EV_Sales_Quantity', data=df,hue='State', palette='bright')
plt.xlabel('States')
plt.ylabel('EV Sales');
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Vehicle_Category',y='EV_Sales_Quantity',data=df, hue='Vehicle_Category',
palette='bright')
plt.title('Analysis by Vehicle Category')
plt.xlabel('Vehicle Category')
plt.ylabel('EV Sales')
plt.xticks(rotation=0);
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Vehicle_Type', y='EV_Sales_Quantity',data=df,hue='Vehicle_Type', palette='bright')
plt.title('Analysis by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('EV Sales')
plt.xticks(rotation=90);
```
